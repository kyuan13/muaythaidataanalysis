```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

```python
# Send a request to the webpage
url = "https://www.onefc.com/events/#past"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#getting links for the events
results = soup.find_all('div', id='past-events-section')
href_links = []
for event_div in results:
    links = event_div.find_all('a', class_='title')
    for link in links:
        print(link.get_text(), link['href'])
        href_links.append(link['href'])
print(href_links)
```

```python
href_links_fighter_one = []
href_links_fighter_two = []
wins_fighter_one = []
wins_fighter_two = []
#getting links for the fighters on the cards
for link in href_links:
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    fighter_one = soup.find_all('a', class_='face face1')
    fighter_two = soup.find_all('a', class_='face face2')
    for fighter in fighter_one:
        print(fighter.get_text(), fighter['href'])
        href_links_fighter_one.append(fighter['href'])
        wins_fighter_one.append(fighter.get_text())
    for fighter in fighter_two:
        print(fighter.get_text(), fighter['href'])
        href_links_fighter_two.append(fighter['href'])
        wins_fighter_two.append(fighter.get_text())
print(href_links_fighter_one)
print(href_links_fighter_two)
print(wins_fighter_one)
print(wins_fighter_two)
cleaned_wins_fighter_one = [item.strip() for item in wins_fighter_one]
print(cleaned_wins_fighter_one)
```

```python
# cleaning the /ns
cleaned_wins_fighter_one = [item.replace('\n', '') for item in wins_fighter_one]
cleaned_wins_fighter_one_df = pd.DataFrame(cleaned_wins_fighter_one,columns=["Wins Fighter 1"])
print(cleaned_wins_fighter_one_df)
cleaned_wins_fighter_two = [item.replace('\n', '') for item in wins_fighter_two]
cleaned_wins_fighter_two_df = pd.DataFrame(cleaned_wins_fighter_two,columns=["Wins Fighter 2"])
```

```python
fighter_one_attributes = pd.DataFrame(columns=["fighter 1 name", "fighter 1 height", "fighter 1 weight", "fighter 1 age"])

for link in href_links_fighter_one:
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')

    # extract name
    name = soup.find('h1', class_="use-letter-spacing-hint my-4")
    fighter_name = name.get_text() if name else "NA"
    
    # extract height
    height_header = soup.find('h5', text='Height')
    if height_header:
        height_value = height_header.find_next('div', class_='value')
        height = height_value.get_text() if height_value else "NA"
    else:
        height = "NA"

    # extract weight
    weight_header = soup.find('h5', text='Weight Limit')
    if weight_header:
        weight_value = weight_header.find_next('div', class_='value')
        if weight_value:
            weight = weight_value.get_text()
            weight_in_lbs = float(weight.split()[0]) if weight else "NA"
        else:
            weight_in_lbs = "NA"
    else:
        weight_in_lbs = "NA"
    
    # extract age
    age_header = soup.find('h5', text='Age')
    if age_header:
        age_value = age_header.find_next('div', class_='value')
        age = age_value.get_text() if age_value else "NA"
    else:
        age = "NA"

    # append
    new_row = pd.DataFrame([[fighter_name, height, weight_in_lbs, age]], columns=["fighter 1 name", "fighter 1 height", "fighter 1 weight", "fighter 1 age"])
    fighter_one_attributes = pd.concat([fighter_one_attributes, new_row], ignore_index=True)


print(fighter_one_attributes)
```

```python
fighter_two_attributes = pd.DataFrame(columns=["fighter 2 name", "fighter 2 height", "fighter 2 weight", "fighter 2 age"])
for link in href_links_fighter_two:
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')

    # extract name
    name = soup.find('h1', class_="use-letter-spacing-hint my-4")
    fighter_name = name.get_text() if name else "NA"
    
    # extract height
    height_header = soup.find('h5', text='Height')
    if height_header:
        height_value = height_header.find_next('div', class_='value')
        height = height_value.get_text() if height_value else "NA"
    else:
        height = "NA"

    # extract weight
    weight_header = soup.find('h5', text='Weight Limit')
    if weight_header:
        weight_value = weight_header.find_next('div', class_='value')
        if weight_value:
            weight = weight_value.get_text()
            weight_in_lbs = float(weight.split()[0]) if weight else "NA"
        else:
            weight_in_lbs = "NA"
    else:
        weight_in_lbs = "NA"
    
    # extract age
    age_header = soup.find('h5', text='Age')
    if age_header:
        age_value = age_header.find_next('div', class_='value')
        age = age_value.get_text() if age_value else "NA"
    else:
        age = "NA"

    # append
    new_row = pd.DataFrame([[fighter_name, height, weight_in_lbs, age]], columns=["fighter 2 name", "fighter 2 height", "fighter 2 weight", "fighter 2 age"])
    fighter_two_attributes = pd.concat([fighter_two_attributes, new_row], ignore_index=True)


print(fighter_two_attributes)
```

```python
combined_df = pd.concat([fighter_one_attributes, fighter_two_attributes, cleaned_wins_fighter_one_df, cleaned_wins_fighter_two_df], axis = 1)
```

```python
import numpy as np
# see which fighter is heavier
combined_df['heavier_fighter'] = combined_df.apply(
    lambda row: 'NA' if row['fighter 1 weight'] == 'NA' or row['fighter 2 weight'] == 'NA' 
                else ('fighter_1' if row['fighter 1 weight'] > row['fighter 2 weight'] else 'fighter_2'), axis=1)
# did the heavier fighter win?
combined_df['heavier_fighter_won'] = combined_df.apply(
    lambda row: np.NaN if row['heavier_fighter'] == 'NA'
    else ((row['heavier_fighter'] == 'fighter_1' and 'WIN' in str(row['Wins Fighter 1'])) or 
                      (row['heavier_fighter'] == 'fighter_2' and 'WIN' in str(row['Wins Fighter 2']))),
    axis=1
)
print(combined_df)
sns.countplot(x='heavier_fighter_won', data=combined_df)
plt.title('Does the Heavier Fighter Usually Win?')
plt.xlabel('Heavier Fighter Won (True/False)')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```
